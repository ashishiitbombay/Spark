###  Exploratory Data Analysis(EDA)
EDA is an important step to build any machine learning project, it is an approach of analyzing datasets to summarize their main characteristics. With the help of EDA, we can able to take knowledge of features by just observing the plots and graphs.

So, here we will use some frequently used visualization techniques for observing data:

### Regional Analysis

```python
#To check in which Region maximum transactions
df['Region'].value_counts().plot.pie(colors=['b','r','g','y'],autopct="%.1f%%")
```

In this pie chart seen that the maximum transaction were made in West Region

### Region wise Sales and Profit

```python
df.groupby(['Region'])['Sales','Profit'].sum().plot.bar(color=['red','blue'],figsize=(12,6))
plt.xlabel("REGION",fontdict={'color':'white','fontsize':15})
plt.title("REGION WISE SALES AND PROFIT",fontdict={'color':'white','fontsize':20})
```

Here bar plot shown highest Sales and profit in  the west region.

####  CUSTOMER ANALYSIS

```python
# To check maximum Sales and Profit in each segment
df.groupby(['Segment'])['Sales','Profit'].sum().plot.bar(color=['green','blue'],figsize=(10,5))
plt.xlabel("SEGMENT",fontdict={'color':'white','fontsize':12})
plt.title("SEGMENT WISE SALES AND PROFIT",fontdict={'color':'white','fontsize':15})
```

Here the bar plot shown that the maximum profit and sales in Consumer segment

### Product Analysis

```python
# To check profit and sales Product wise
df.groupby(['Category'])['Sales','Profit'].sum().plot.bar(color=['#FF8C00','#3CB371'],figsize=(12,6))
plt.xlabel("CATEGORY",fontdict={'color':'white','fontsize':15})
plt.title("PRODUCT CATEGORY WISE SALES AND PROFIT",fontdict={'color':'white','fontsize':20})
```

Here the barplot shown that the highest profit and seles in Technology.

```python
#Checking Ship Mode Segment wise
plt.subplots(figsize=(16,8))
sns.countplot(x='Segment',hue='Ship Mode',data=df)
plt.xlabel('SEGMENT',fontdict={'color':'white','fontsize':15})
plt.title("SEGMENT WISE SHIP MODE",fontdict={'color':'white','fontsize':15})
```

Here barplot shown that in each segment most of the transaction has been shipped under Standard Class.

###  TOP PRODUCTS

```python
# Now we will check the Top Products Sold
df.groupby(['Sub-Category'])['Sales'].sum().sort_values(ascending=False).plot.bar(color=['blue'],figsize=(12,6))
plt.xlabel("SUB CATEGORY",fontdict={'color':'white','fontsize':15})
plt.ylabel("SALES",fontdict={'color':'white','fontsize':15})
plt.title("SUB-CATEGORY WISE SALES",fontdict={'color':'white','fontsize':20})
```

With this we concluded that Phones,Chairs ,Storage,Tables and Binders are being sold the most. Whereas Fasteners,Labels and Envelopes were sold the least

### CITY ANALYSIS

```python
#To check top 5 cities based on Sales
df.groupby(['City'])['Sales'].sum().sort_values(ascending=False).head().plot.bar(color=['red'],figsize=(10,5))
plt.xlabel("CITY",fontdict={'color':'white','fontsize':12})
plt.ylabel("SALES",fontdict={'color':'white','fontsize':12})
plt.title("CITY WISE SALES",fontdict={'color':'white','fontsize':15})
```

so, from the plot New York City has highest sales followed by Los Angeles and Seattle whereas lowest sales in Philadelphia

```python
#To check the profit earned from the top 5 cites
df.groupby(['City'])['Profit'].sum().sort_values(ascending=False).head().plot.bar(color=['blue'],figsize=(10,5))
plt.xlabel("CITY",fontdict={'color':'white','fontsize':12})
plt.ylabel("PROFIT",fontdict={'color':'white','fontsize':12})
plt.title("CITY WISE PROFIT",fontdict={'color':'white','fontsize':15})
```

From the plot we seen that highest profit in New york city whereas lowest profit in Detroit

###  Correlation plot

The correlation plot shows the correlation coefficient between variables. This plot contains the correlation matrix-like table.

```python
plt.figure(figsize=[17,8])
sns.heatmap(df.corr(),annot=True)
```

From above Heatmap we concluded there is a Negative correlation in between Profit and Discount whereas, a Positive correlation between Profit and Sales
